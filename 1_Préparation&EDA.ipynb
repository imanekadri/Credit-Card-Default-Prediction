```python
TARGET = "default.payment.next.month"
assert TARGET in data.columns, "Target column not found!"
```

```python
data.duplicated().sum()
```

```python
cat_cols = ["SEX", "EDUCATION", "MARRIAGE"]
```

```python
cat_cols
```

```python
num_cols = data.select_dtypes(include=["int64", "float64"]).columns.tolist()
```

```python
num_cols
```

```python
num_cols = [x for x in num_cols if x not in cat_cols]
num_cols = [c for c in num_cols if c not in [TARGET]]
```

```python
num_cols
```

```python
for i in num_cols:
    plt.figure(figsize=(6,3))
    plt.hist(data[i], bins=25)
    plt.title(f'Histogram of {i}')
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(
    x='SEX',
    hue='default.payment.next.month',
    data=data
)

plt.title("Sex vs Default Payment")
plt.xlabel("Sex (1 = Male, 2 = Female)")
plt.ylabel("Number of Clients")
plt.legend(title="Default", labels=["No Default", "Default"])
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x='default.payment.next.month',
    y='AGE',
    data=data
)

plt.title("Age vs Default Payment")
plt.xlabel("Default (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.hist(data[data['default.payment.next.month']==0]['AGE'],
         bins=20, alpha=0.6, label='No Default')

plt.hist(data[data['default.payment.next.month']==1]['AGE'],
         bins=20, alpha=0.6, label='Default')

plt.title("Age Distribution by Default")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x='default.payment.next.month',
    y='PAY_0',
    data=data
)

plt.title("PAY_0 vs Default Payment")
plt.xlabel("Default (0 = No, 1 = Yes)")
plt.ylabel("Payment Delay (PAY_0)")
plt.show()
```

```python
corr = data[num_cols].corr()
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
